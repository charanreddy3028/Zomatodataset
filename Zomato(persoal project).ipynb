# ZOMATO BANGLORE RESTAURANTS DATASET


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading CSV File

```python
df=pd.read_csv('zomato.csv')
```

```python
df.columns
```

## Dropping of not useful colums

```python
df=df.drop(['url','address','reviews_list','menu_item','phone','listed_in(city)'],axis=1)
```

```python
df
```

## Checking null values

```python
df.info()
```

# Removing Duplicates

```python
df.drop_duplicates(inplace=True)
df
```

```python
df['rate'].unique()
```

## Converting the rate column into float

```python
def null_val(value):
    if(value=='NEW' or value == '-'):
        return np.nan
    else:
        value= str(value).split('/')
        value=value[0]
        return float(value)
    
df['rate']=df['rate'].apply(null_val)
df['rate']
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df['rate']
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

## Cleaning of dish_liked

```python
df=df.drop('dish_liked',axis=1)
```

```python
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df
```

```python
df.info()
```

## Renaming the column names

```python
df.rename(columns={'approx_cost(for two people)':'2_plates_cost','listed_in(type)':'Type'},inplace=True)
```

```python
df
```

```python
df['location'].unique()
```

## Cleaning the 2platecost column

```python
df['2_plates_cost'].unique()
```

```python
def coma_val(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['2_plates_cost']=df['2_plates_cost'].apply(coma_val)
df['2_plates_cost'].unique()
```

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
R_type=df['rest_type'].value_counts(ascending=False)
R_type
```

```python
# rest_type lessthan 1000 considered as others
```

```python
R_type_lessthan_1000=R_type[R_type <1000]
R_type_lessthan_1000
```

## restaurants having  rest_type count is less than 1000 in frequency as others

```python
def r_type(value):
    if value in R_type_lessthan_1000:
        return "others"
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(r_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
location=df['location'].value_counts(ascending=False)
```

```python
# resautants in less than 500 locations
```

```python
location_lessthan500=location[location<500]
location_lessthan500
```

## locations less than 500 in frequency as others

```python
def location_val(value):
    if value in location_lessthan500:
        return "Others"
    else:
        return value
    
df['location']=df['location'].apply(location_val)
df['location'].value_counts(ascending= True)
```

```python
df.head()
```

```python
cuisines=df['cuisines'].value_counts(ascending=True)
cuisines
```

```python
cuisines_lessthan150=cuisines[cuisines<150]
cuisines_lessthan150
```

```python
def cuise_val(value):
    if value in cuisines_lessthan150:
        return "Others"
    else:
        return value
    
df['cuisines']=df['cuisines'].apply(cuise_val)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts(ascending=True)
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
print('Enter your choice')
print('1.Add')
print('2.Sub')
print('3.Mul')
print('4.Div')

choice=input('Enter Your choice:')
num1=float(input('Enter 1st Number:'))
num2=float(input('Enter 2nd Number'))

if choice=='1':
    print(num1,'+',num2,'=',add(num1,num2))
elif choice=='2':
    print(num1,'-',num2,'=',sub(num1,num2))
elif choice=='3':
    print(num1,'*',num2,'=',mul(num1,num2))
elif choice=='4':
    print(num1,'/',num2,'=',div(num1,num2))
else:
    print('Invalid Input')

```
